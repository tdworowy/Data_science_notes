```python
%matplotlib inline
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.Graph()
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(3,1)
G.add_edge(3,4)
G.add_edge(4,1)
G.add_edge(4,2)
nx.draw_networkx(G)
plt.show()
```

```python
G = nx.Graph()
G.add_edges_from([(1,2), (2,3),(3,4), (4,1), (5,4)])
nx.draw_networkx(G)
plt.show()
```

```python
k = nx.fast_gnp_random_graph(15,0.40)
nx.draw_networkx(k)
plt.show()
```

```python
G = nx.krackhardt_kite_graph()
nx.draw_networkx(G)
plt.show()
```

```python
print(nx.has_path(G, source=1, target=9))
print(nx.shortest_path(G, source=1, target=9))
print(nx.shortest_path_length(G, source=1, target=9))
```

```python
all_paths = list(nx.shortest_simple_paths(G, source=1, target=9))
for path in all_paths:
    print(path)
```

```python
paths = nx.all_pairs_shortest_path(G)
for path in paths:
    print(path)
```
